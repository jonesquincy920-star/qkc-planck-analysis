# 02 — Spot Analysis
Extract the patch around the QKC target (l=−57°, b=−27°) and characterise it.

```python
import sys, json
sys.path.insert(0, '../src')
from qkc import load_map, extract_patch, patch_stats, plot_patch
import healpy as hp
import numpy as np
```

```python
cmap = load_map('../data/COM_CMB_IQU-smica_2048_R3.00_full.fits', field=0, nside_out=512)
```

```python
# Extract 10-degree disc around QKC target
ipix, vals = extract_patch(cmap, l_deg=-57.0, b_deg=-27.0, radius_deg=10.0)
print(f'Patch pixels: {len(ipix)}')
stats = patch_stats(vals)
print(json.dumps(stats, indent=2))
```

```python
nside = hp.get_nside(cmap)
fig = plot_patch(ipix, vals, nside, title='QKC Target Patch (l=−57°, b=−27°)')
fig.savefig('../data/qkc_patch.png', dpi=150, bbox_inches='tight')
```
