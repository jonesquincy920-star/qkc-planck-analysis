# 01 — Data Loading
Download the Planck 2018 SMICA CMB map and inspect it.

```python
import sys
sys.path.insert(0, '../src')
from qkc import download_planck_map, load_map
import healpy as hp
import numpy as np
```

```python
# Download SMICA map (large file ~500 MB; cached after first run)
map_path = download_planck_map(method='SMICA')
print('Map saved to:', map_path)
```

```python
# Load at reduced resolution for fast exploration (nside=256)
cmap = load_map(map_path, field=0, nside_out=256)
print(f'nside={hp.get_nside(cmap)}, npix={len(cmap)}')
print(f'T range: {cmap.min():.2f} to {cmap.max():.2f} μK')
```

```python
from qkc import plot_mollweide
fig = plot_mollweide(cmap, title='Planck 2018 SMICA — Temperature', mark_qkc=True)
fig.savefig('../data/fullsky_smica.png', dpi=150, bbox_inches='tight')
```
